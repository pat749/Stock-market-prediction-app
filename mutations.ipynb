```python
from models.dnn_regression import DenseNeuralNetwork
```

```python
import json
```

```python
parent_model = DenseNeuralNetwork.random_models(1, "dense")[0]
print("Parent model model options:")
print(json.dumps(parent_model.model_options, indent=4))
for mutation in DenseNeuralNetwork.MUTATIONS["dense"]:
    print("Mutation:", mutation)
    model = DenseNeuralNetwork(
        DenseNeuralNetwork.evolve_model_options(parent_model.model_options, mutation),
        parent_model.input_options,
        parent_model.stock_code
    )
    print(json.dumps(model.model_options, indent=4))
```

```python
parent_model_with_layers = DenseNeuralNetwork(
    DenseNeuralNetwork.evolve_model_options(parent_model.model_options, "add_dense_layer"),
    parent_model.input_options,
    parent_model.stock_code
)
print("Parent model with layers model options:")
print(json.dumps(parent_model_with_layers.model_options, indent=4))
for mutation in DenseNeuralNetwork.MUTATIONS["dense"]:
    print("Mutation:", mutation)
    model = DenseNeuralNetwork(
        DenseNeuralNetwork.evolve_model_options(parent_model_with_layers.model_options, mutation),
        parent_model_with_layers.input_options,
        parent_model_with_layers.stock_code
    )
    print(json.dumps(model.model_options, indent=4))
```
